```python
file_name_close = 'daily_close_prices.csv'
file_name_high = 'daily_high_prices.csv'
file_name_low = 'daily_low_prices.csv'
file_name_open = 'daily_open_prices.csv'
file_name_key = 'stock_info.csv'
```
